### MatplotLib Basicos
Simple graph plotting=>

```python
import matplotlib.pyplot as plt
import numpy as np

 # 100 values within range non-inclusive
 
 #keep diff smaller to make more smooth
x=np.arange(0,10,.1)
y=np.sin(x)
p=np.linspace(0,10,100)
q=np.sin(p)



plt.plot(x,y,color='green',linestyle='--',label='Sine', linewidth=2)
plt.grid()
plt.title("Sine wave")
plt.legend()
plt.fill_between(x,y,color='skyblue',alpha=0.8)

plt.xlabel("value of angle")
plt.ylabel("Sin value of correspondednt angle")
plt.show()
```

### Multiple plot or Subplots

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x),label='sin(x)')
plt.title("Sine")

plt.plot(x, np.cos(x),label='cos(x)')
plt.title("Cosine")

plt.legend()

plt.plot()

```

### Subplots (multiple )

```python
# returns Figure,Axes array
fig,axs=plt.subplots(2,1,figsize=(8,4))
#2 rows 1 col

x=np.arange(1,10,.1)
y=np.sin(x)
y2=np.cos(x)


axs[0].plot(x,y)
axs[0].set_title('Sine wave')

axs[1].plot(x,y2)
axs[1].set_title('Cos wave')



# plt.legend()
plt.tight_layout() #good shi
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(8,4))
# horizontal stacking
# axs is 1d array bruh !?

x=np.linspace(1,5,100)

y=np.sin(x)
y2=np.cos(x)

axs[0].plot(x,y)
axs[0].set_title("Sine wave")
axs[0].set_ylabel("Sine")

axs[1].plot(x,y2)
axs[1].set_title("Cos wave")
axs[1].set_ylabel("Cos")


plt.tight_layout()
plt.savefig("horizontalStack.png",dpi=400,transparent=True)
plt.show() # it cclears out so save first man

# saving plot as png
```

### Scatter/Bar/Charts

```python
x=np.arange(1,100,2)
y=np.cos(x)
# must be same size

# just show individual points 
color=np.random.rand(len(x))
plt.scatter(x,y,c=color,cmap='viridis')
#scatter with color grad
plt.show()

```

```python
# bar chart 
subjects=['Math','Eng','CompSci','Econ']
marks=[80,67,58,74]

plt.bar(subjects,marks)
plt.show()
```

```python
#histograms
data=np.random.randn(500)
data2=np.arange(1,500,1)

plt.hist(data,bins=30) #number of bars
plt.title('Histogram')
plt.grid()
plt.show()
```
